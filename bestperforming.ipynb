```python
import os
import torch
import requests
from collections import Counter
from PIL import Image
from torchvision import transforms
from tqdm import tqdm
from sklearn.metrics import accuracy_score
import random
import json

# Path to the directory where images are stored
image_folder_path = 'IRFL_images/images'
# IRFL_idiom_detection_task = load_dataset("lampent/IRFL", 'idiom-detection-task')["test"]

# Define the data augmentation transformations
data_transforms = transforms.Compose([
    transforms.Resize((224, 224)),  # Resize images to 224x224 pixels
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.ColorJitter(brightness=0.1, contrast=0.1, saturation=0.1, hue=0.1),
    transforms.ToTensor(),
])

# Get the image path for the given image name
def get_image_path(image_name):
    return os.path.join(image_folder_path, f"{image_name}.jpeg")

# Load the image from the file and apply transformations
def get_image(image_name):
    image_path = get_image_path(image_name)
    if os.path.exists(image_path):
        img = Image.open(image_path).convert("RGB")  # Ensure image is in RGB format
        img = data_transforms(img)  # Apply data augmentation
        return img
    else:
        print(f"Image file not found: {image_path}")
        return None

# Preprocess the image and get its CLIP vector
def get_clip_img_vector(img):
    if img is None:
        return None
    cue_clip_img = img.unsqueeze(0).to(device)  # Move to device
    with torch.no_grad():
        cue_clip_img_encoded = model.encode_image(cue_clip_img)
    cue_clip_img_encoded /= cue_clip_img_encoded.norm(dim=-1, keepdim=True)
    return cue_clip_img_encoded

# Calculate the similarity between two vectors
def get_vectors_similarity(v1, v2):
    return v1.detach().cpu().numpy() @ v2.detach().cpu().numpy().T

# Tokenize the text and get its CLIP vector
def get_clip_txt_vector(text):
    cue_clip_txt = clip.tokenize([text]).to(device)
    with torch.no_grad():
        cue_clip_txt_encoded = model.encode_text(cue_clip_txt)
    cue_clip_txt_encoded /= cue_clip_txt_encoded.norm(dim=-1, keepdim=True)
    return cue_clip_txt_encoded

# ConceptNet API to fetch relations for a phrase
def get_conceptnet_relations(phrase):
    # Construct the URL for the ConceptNet API
    url = f"http://api.conceptnet.io/c/en/{phrase.lower()}"

    try:
        # Make the API request
        response = requests.get(url)

        # Check if the response was successful (status code 200)
        if response.status_code != 200:
            print(f"Error fetching data from ConceptNet: {response.status_code}")
            return []  # Return an empty list in case of an error

        # Attempt to parse the response as JSON
        try:
            response_json = response.json()
        except ValueError:
            print(f"Invalid JSON received from ConceptNet for phrase: {phrase}")
            return []  # Return empty list for invalid JSON

        # Extract relations if the response contains edges
        relations = []
        for edge in response_json.get('edges', []):
            relations.append({
                "start": edge['start']['label'],
                "relation": edge['rel']['label'],
                "end": edge['end']['label'],
                "weight": edge['weight']
            })
        return relations

    except requests.exceptions.RequestException as e:
        print(f"Error during requests to ConceptNet: {e}")
        return []  # Return empty list on request errors

# Calculate similarity score between a phrase and an image, integrating ConceptNet
def get_phrase_image_similarity_score(phrase, image_name, definitions):
    # Fetch ConceptNet relations for the phrase
    conceptnet_relations = get_conceptnet_relations(phrase)

    # Extract the related concepts from ConceptNet, if any
    related_concepts = " ".join([rel['end'] for rel in conceptnet_relations])

    # Prepare the phrase with additional ConceptNet relations
    phrase_with_conceptnet = phrase + " " + related_concepts

    # If definitions exist, combine them with the phrase and related concepts
    if definitions:
        definition_prompt = '.'.join(definitions) + '.'
        phrase_with_conceptnet += '.' + definition_prompt

    # Encode the enriched phrase
    phrase_clip_txt_encoded = get_clip_txt_vector(phrase_with_conceptnet)

    # Load the image and encode it
    img = get_image(image_name)
    if img is None:
        return 0  # Default value for missing image
    clip_cand_img_encoded = get_clip_img_vector(img)
    if clip_cand_img_encoded is None:
        return 0  # Default value for error

    # Compute the similarity between the phrase and image vectors
    cand_txt_img_sim = get_vectors_similarity(phrase_clip_txt_encoded, clip_cand_img_encoded).item()
    return cand_txt_img_sim

# Solves the detection task instance for a phrase and its image candidates
def solve_IRFL_detection_task_instance(phrase, candidates, definitions):
    sim_for_image = {}
    for img_name in candidates:
        sim_for_image[img_name] = get_phrase_image_similarity_score(phrase, img_name, definitions)

    if sim_for_image:
        clip_prediction = Counter(sim_for_image).most_common()[0][0]
        return clip_prediction
    else:
        return None

# Preprocesses the task instance before solving
def pre_processing(task_instance):
    answer = json.loads(task_instance['answer'])
    distractors = json.loads(task_instance['distractors'])
    definitions = json.loads(task_instance.get('definition', '[]'))
    candidates = answer + distractors
    random.shuffle(candidates)
    return task_instance['phrase'], answer[0], distractors, candidates, definitions

# Solves the IRFL detection task in batches
def solve_IRFL_detection_task(df, batch_size=16):
    ground_truth = []
    predictions = []

    for start in tqdm(range(0, df.shape[0], batch_size)):
        end = min(start + batch_size, df.shape[0])
        batch_df = df.iloc[start:end]

        for _, row in batch_df.iterrows():
            phrase, answer, distractors, candidates, definitions = pre_processing(row)
            prediction = solve_IRFL_detection_task_instance(phrase, candidates, definitions)

            ground_truth.append(answer)
            predictions.append(prediction)

    accuracy = int(accuracy_score(ground_truth, predictions) * 100)
    return ground_truth, predictions, accuracy
```
